### Merging and Concatenating Data ###

```python
noc = pd.read_csv("noc_regions.csv")
```

```python
noc.head()
```

```python
#left join example of joining two tables of df and noc 
# joining on born_country and NOC

df_new = pd.merge(df, noc, left_on = 'born_country', right_on = 'NOC', how='left')
df_new.head()
```

```python
#change the name 

df_new.rename(columns = {'region':'born_country_full'}, inplace=True)
```
